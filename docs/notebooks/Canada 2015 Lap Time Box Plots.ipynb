# Canada 2015 Formula 1 Lap Times
---
Uses data posted by FIA in PDF format. FormulaPy includes parses to extract this data into column-oriented format (Pandas DataFrame).

## Setup

```python
import sys
import os
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline

rel_path = '../../'

# get access to formulapy module
sys.path.append(os.path.realpath(rel_path))  
```

```python
from formulapy.data.fia.parsers import parse_laptimes
from formulapy.plots import time_box_plot
```

## Load FP1, FP2, FP3 Data from FIA

```python
df_fp1 = parse_laptimes(os.path.realpath(rel_path + 'data/fp1_lap_times_2.pdf'))
df_fp2 = parse_laptimes(os.path.realpath(rel_path + 'data/fp2_lap_times_1.pdf'))
df_fp3 = parse_laptimes(os.path.realpath(rel_path + 'data/fp3_lap_times_2.pdf'))

# show a peek at one of the dataframes
df_fp1.head()
```

## Create a Box Plot for Each Practice Session
Note, lap times greater than 125 seconds are filtered out to provide real lap times. Additional filtering is applied based on each drivers distribution of lap times to try to catch in/out laps.

### FP1

```python
time_box_plot(df_fp1.ix[df_fp1['time'] < 125, :], time_field='time', id_field='name', pit_laps=False)
```

### FP2

```python
time_box_plot(df_fp2.ix[df_fp2['time'] < 125, :], time_field='time', id_field='name', pit_laps=False)
```

### FP3

```python
time_box_plot(df_fp3.ix[df_fp3['time'] < 125, :], time_field='time', id_field='name', pit_laps=False)
```
