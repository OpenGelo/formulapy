```python
con = MySQLdb.connect(host="127.0.0.1", user="ipython", passwd="ipythonnb", db="formula1")
```

```python
laptimes2014 = pd.read_sql("SELECT races.raceId, year, circuitId, races.name AS race_name, laptimes.driverId, lap, position, laptimes.time, \
laptimes.milliseconds, driverRef \
FROM races \
JOIN laptimes \
ON races.raceId=laptimes.raceId \
JOIN drivers \
ON laptimes.driverId=drivers.driverId \
WHERE races.year=2014", con)
```

```python
laptimes2014.head()
```

```python
driver_times = laptimes2014.groupby(['race_name','driverRef'])
```

```python
driver_times.mean()
```

```python
driver_times_min = driver_times.min()
```
